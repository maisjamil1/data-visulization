```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
happiness= pd.read_csv('datasets_894_813759_2016.csv')
happiness.head(5)
# len(happiness['Country'])
```

```python
X = happiness['Happiness Score'].values
y = happiness['Economy (GDP per Capita)'].values
```

```python
sns.set()
```

## How Happiness Score is related to Economy

```python
sns.relplot(x='Happiness Score', y='Economy (GDP per Capita)', data=happiness);
```

```python
graph = sns.barplot(x="Country", y="Happiness Rank", data=happiness, ci=None)
graph.set_xticklabels(labels=list(happiness['Country']));


```

## Happiness scors for each country

```python
height = X[:30]
bars = happiness['Country'].head(30)
y_pos = np.arange(len(bars))
 
plt.bar(y_pos, height)
 
plt.xticks(y_pos, bars, rotation=90)
plt.show()
```

## Health (Life Expectancy)

```python

first_20=happiness['Country'].head(20)
# first_20
f20=happiness['Health (Life Expectancy)'].head(20)

# n = 20
# Z = np.random.uniform(0,1,n)
# plt.pie(Z)
# plt.show()
```

```python
Country_li=[]
for i in first_20:
    Country_li.append(str(i))
# li
Economy_li2=[]
for i in f20:
    Economy_li2.append(float(i))
# Economy_li2

df = pd.DataFrame(Economy_li2, index=Country_li, columns=['x'])
 
# make the plot
df.plot(kind='pie', subplots=True, figsize=(8, 8))

```
